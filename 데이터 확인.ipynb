## 데이터 전처리

```python
batch_size = 64
IMG_SIZE = 224

train_dir = 'VOC2012/dataset'
```

```python
data_info = dict()
x = np.arange(20)
for cls in os.listdir(train_dir):
    data_info[cls] = len(os.listdir(f"{train_dir}/{cls}"))
```

```python
plt.figure(figsize=(18,8))
plt.title('VOC2012 num of classes (only one object)', fontsize=20)
plt.bar(x, list(data_info.values()))
plt.xticks(x, list(data_info.keys()), fontsize=15, rotation=45)
plt.ylabel('The number of classes', fontsize=20)
plt.xlabel('The kind of classes', fontsize=20)

for i, v in enumerate(x):
    plt.text(v, list(data_info.values())[i], list(data_info.values())[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 12, 
             color='blue',
             horizontalalignment='center',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)

plt.show()
```
