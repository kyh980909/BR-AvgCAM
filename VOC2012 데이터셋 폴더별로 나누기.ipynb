```python
label_list_path = 'VOC2012/Annotations'
image_data_list_path = 'VOC2012/JPEGImages'
dataset_list_path = 'VOC2012/dataset'
class_list = ['aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus', 'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike', 'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tvmonitor']
```

```python
def get_label(xml):
    label = xml.find('object').find('name').text
    num = len(xml.findall('object'))
    p_size = xml.find('size')
    p_box = xml.find('object').find('bndbox')
    size = {'width':int(float(p_size.find('width').text)),'height': int(float(p_size.find('height').text))}
    box = {'xmin':int(float(p_box.find('xmin').text)), 'ymin' : int(float(p_box.find('ymin').text)),'xmax': int(float(p_box.find('xmax').text)),'ymax': int(float(p_box.find('ymax').text))}
    xmin, ymin, xmax, ymax = box['xmin'] / size['width'] * 224, box['ymin'] / size['height'] * 224, box['xmax'] / size['width'] * 224,box['ymax'] / size['height'] * 224
    w, h = xmax - xmin, ymax - ymin
    return {'xmin':xmin, 'ymin':ymin, 'xmax':xmax,'ymax':ymax,'w':w, 'h':h, 'label':label, 'object_num' : num}
```

```python
for cls in class_list:
    os.mkdir(f"{dataset_list_path}/{cls}")
```

```python
for data in os.listdir(image_data_list_path):
    label = get_label(parse(f'{label_list_path}/{data.replace("jpg","xml")}'))
    if label['object_num'] == 1:
        shutil.copy2(f'{image_data_list_path}/{data}',f'{dataset_list_path}/{label["label"]}/{data}')
        print(label['label'])
        print(f'{image_data_list_path}/{data} -> {dataset_list_path}/{label["label"]}/{data}')
```
